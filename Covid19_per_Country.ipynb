```python
df["death_rate"] = df['Deaths'] / df['Confirmed']
df['recovery_rate'] = df['Recovered'] / df['Confirmed']
df['active_cases'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
```

```python
df[['death_rate', 'recovery_rate', 'active_cases']]
```

```python
## Checking if there is Nan Values
death_rate_Nan = df['death_rate'].isna().sum()
recovery_rate_Nan = df['recovery_rate'].isna().sum()
active_cases_Nan = df['active_cases'].isna().sum()

print(death_rate_Nan)
print(recovery_rate_Nan)
print(active_cases_Nan)
```

```python
# filtering countries with greater than 10000 confirmed cases
filtered_df = df[df['Confirmed'] >= 10000]
filtered_df.head()
```

```python
# Checking the top 10 countries with highest death rate
top_10_deaths = filtered_df.sort_values(by="death_rate", ascending=False)[['Country/Region', 'death_rate']].head(10)

# Checking the top 10 countries with highest recovery rate
top_10_recovery = filtered_df.sort_values(by="recovery_rate", ascending=False)[['Country/Region', 'recovery_rate']].head(10)

# Checking the top 10 countries with highest active cases
top_10_active = filtered_df.sort_values(by='active_cases', ascending=False)[['Country/Region', 'active_cases']].head(10)

# Checking the top 10 countries with highest rate increase per week
top_10_rate_week = filtered_df.sort_values(by='1 week % increase', ascending=False)[['Country/Region', '1 week % increase']].head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Visualize
plt.figure(figsize=(10,6))
sns.barplot(x='death_rate', y='Country/Region', data=top_10_deaths, palette = 'Reds_r')
plt.xlabel('Death Rate')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries with Highest Death Rate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='recovery_rate', y='Country/Region', data=top_10_recovery, palette = 'Greens')
plt.xlabel('Recovery Rate')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries with Highest Recovery Rate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='active_cases', y='Country/Region', data=top_10_active, palette = 'Oranges')
plt.xlabel('Active Cases')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries with Highest Active Cases')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='1 week % increase', y='Country/Region', data=top_10_rate_week, palette = 'Blues')
plt.xlabel('1-Week % Increase')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries with Highest Active Cases')
plt.show()
```

```python
heatmap_data = df[['Confirmed', 'Deaths', 'Active', 'New cases']]
```

```python
# Global Confirmed, Global Deaths, Global Recovered

global_confirmed = df['Confirmed'].sum()
global_deaths = df['Deaths'].sum()
global_recovered = df['Recovered'].sum()

print('Global Confirmed:', global_confirmed)
print('Global Deaths:', global_deaths)
print('Global Recovered:', global_recovered)
```

```python
# Global Death Rate
global_death_rate = global_deaths / global_confirmed
global_recovery_rate = global_recovered / global_confirmed

print(f"Global Death Rate: {global_death_rate}")
print(f"Global Recovery Rate: {global_recovery_rate}")
```
